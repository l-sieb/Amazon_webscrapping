```python
# import libraries

from bs4 import BeautifulSoup as BS
import requests as rq
import time
import datetime
```

```python
# Connect to website
URL = 'https://e-benedetti.fr/shampoing-solide-pachamamai-sweetie-demelant-65g.html'


headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

page = rq.get(URL, headers=headers)
soup1 = BS(page.content, "html.parser")
soup2 = BS(soup1.prettify(), 'html.parser')
soup2
```

```python

title = soup2.find(itemprop="name").get_text()
price = soup2.find(id="our_price_display").get_text()
rating = soup2.find(class_="left text-size-07").get_text()

print(title)
print(price)
print(rating)
```

```python
# clean
title = title.strip()

rating = str(rating)
rating = rating.strip()

price = str(price)
price = price.strip()
```

```python
print(title)
print(rating)
print(price)
```

```python
# Data stamp to know when I collected the data

today = datetime.date.today()
print(today)
```

```python
import csv

header = ['Title','Price','Ratings','Date']
data = [title, price, rating, today]

with open('Ecommerce_WebScrapper_Dataset.csv','w',newline='',encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(data)
```

```python
# Automate
def check_price():
    URL = 'https://e-benedetti.fr/shampoing-solide-pachamamai-sweetie-demelant-65g.html'


    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

    page = rq.get(URL, headers=headers)
    soup1 = BS(page.content, "html.parser")
    soup2 = BS(soup1.prettify(), 'html.parser')
    
    title = soup2.find(itemprop="name").get_text()
    price = soup2.find(id="our_price_display").get_text()
    rating = soup2.find(class_="left text-size-07").get_text()

    title = title.strip()
    rating = rating.strip()
    price = price.strip()
    
    today = datetime.date.today()
    
    header = ['Title','Price','Rating','Date']
    data = [title, price, rating, today]
    
    with open('Ecommerce_WebScrapper_Dataset.csv','w',newline='',encoding='UTF8') as f:
        writer = csv.writer(f)
        writer.writerow(data)
        
# This is everything we need and now it has to be on a timer (every 24h)
```

```python
while(True):
    check_price()
    time.sleep(86400)
```
